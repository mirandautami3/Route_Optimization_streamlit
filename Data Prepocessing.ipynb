```python
import pandas as pd

# Load data dari CSV
df = pd.read_csv('surabaya_routes.csv')

# Menampilkan 5 baris pertama untuk melihat struktur data
print(df.head())
```

```python
# Menghapus duplikat jika ada
df = df.drop_duplicates()

# Menampilkan data setelah duplikat dihapus
print(df.head())
```

```python
print(f"Jumlah data sebelum menghapus duplikat: {len(df)}")
df = df.drop_duplicates()
print(f"Jumlah data setelah menghapus duplikat: {len(df)}")
```

```python
# Memeriksa missing values
print(df.isnull().sum())

# Jika ada missing values, kita bisa menghapusnya atau mengisinya
df = df.dropna()  # Menghapus baris dengan nilai yang hilang
# atau
# df = df.fillna(0)  # Mengisi nilai yang hilang dengan 0

```

```python
# Menampilkan statistik deskriptif untuk kolom numerik
print(df.describe())

# Menampilkan distribusi data
import matplotlib.pyplot as plt

# Histogram untuk Distance_km
plt.hist(df['Distance_km'], bins=20, edgecolor='black')
plt.title('Distribusi Jarak (km)')
plt.xlabel('Jarak (km)')
plt.ylabel('Frekuensi')
plt.show()

# Histogram untuk Duration_min
plt.hist(df['Duration_min'], bins=20, edgecolor='black')
plt.title('Distribusi Waktu Tempuh (menit)')
plt.xlabel('Waktu Tempuh (menit)')
plt.ylabel('Frekuensi')
plt.show()

```

```python
# Menyimpan data yang sudah diproses
df.to_csv('surabaya_routes_clean.csv', index=False)

# Menampilkan 5 baris pertama untuk memverifikasi hasil
print(df.head())
```
